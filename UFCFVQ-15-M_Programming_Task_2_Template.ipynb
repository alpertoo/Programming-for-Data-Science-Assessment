### Requirement FR8 - Detect and remove any outliers in the data used for your "interesting" pattern or trend

I used boxplots to detect outliers. Boxplots provide simple but effective graphics to look at the summary of the distribution of the data. Boxplots visualise summaries with the 20th, 50th and 75th percentiles and makes it easy to get ideas about the data by just looking at the visualisation. Boxplots visualise the Inter Quartile Range. Quartiles are three data points dividing the data into four equally divided parts. These parts represent the quarters of the data. Q1 is the number in the middle between the median and the smallest number of the data set. Q2 represent the median. Q3 is the number in the middle between the median and the highest value. We can calculate the Inter Quartile Range by substracting the first quartile from the third quartile.

There are *no outlier values* for these disorders.

#### Boxplot for Bipolar disorder and Eating disorders

```python
columns_to_choose = ["Bipolar disorder (%)", "Eating disorders (%)"]
world_data_subset = world_data[columns_to_choose]
plt.figure()
sns.boxplot(world_data_subset)
```

#### Inter Quartile Range calculation for bipolar disorder

```python
# IQR
Q1 = np.percentile(world_data["Bipolar disorder (%)"], 25, method="midpoint")
Q3 = np.percentile(world_data["Bipolar disorder (%)"], 75, method="midpoint")
IQR = Q3 - Q1
print("IQR: ", IQR)

# Above Upper bound
upper = Q3 + 1.5 * IQR
upper_array = np.array(world_data["Bipolar disorder (%)"] >= upper)
print("Upper Bound:", upper)
print("Total values over upper bound: ",upper_array.sum())
 
# Below Lower bound
lower = Q1 - 1.5 * IQR
lower_array = np.array(world_data["Bipolar disorder (%)"] <= lower)
print("Lower Bound:", lower)
print("Total values below lower bound: ",lower_array.sum())
```

#### Inter Quartile Range calculation for eating disorders

```python
# IQR
Q1 = np.percentile(world_data["Eating disorders (%)"], 25, method="midpoint")
Q3 = np.percentile(world_data["Eating disorders (%)"], 75, method="midpoint")
IQR = Q3 - Q1
print("IQR: ", IQR)

# Above Upper bound
upper = Q3+ 1.5 *IQR
upper_array = np.array(world_data["Eating disorders (%)"] >= upper)
print("Upper Bound:", upper)
print("Total values over upper bound: ",upper_array.sum())
 
# Below Lower bound
lower = Q1-1.5*IQR
lower_array = np.array(world_data["Eating disorders (%)"] <= lower)
print("Lower Bound:", lower)
print("Total values below lower bound: ",lower_array.sum())
```

### Requirement FR9 - Define a hypothesis to test your â€œinterestingâ€ pattern or trend

*Question: Is there a statistically significant relation between the prevalence of Bipolar disorder and Eating disorders.*

The null hypothesis suggests that there is no underlying relationship between the prevalence rates of Bipolar disorder and Eating disorders. The alternative hypothesis says that there is a positive correlation.

#### Null hypothesis
<div align="center"><font size=2.5><b><i>H</i><sub>0</sub>: there is no relation between Bipolar disorder and Eating disorders.</b></font></div>

#### Alternative hypothesis
<div align="center"><font size=2.5><b><i>H</i><sub>a</sub>: there is a relation between Bipolar disorder and Eating disorders.</b></font></div>

### Requirement FR10 - Test your hypothesis with statistical significance level of 0.05

```python
bipolar_disorder = world_data["Bipolar disorder (%)"]
eating_disorders = world_data["Eating disorders (%)"]

# The Pearson correlation test
corr_coefficient, p_value = stats.pearsonr(bipolar_disorder, eating_disorders)

print(f"Pearson Correlation Coefficient: {corr_coefficient}")
print(f"P-value: {p_value}")

# Check if the p-value is lower than 0.05
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant correlation.")
else:
    print("Fail to reject the null hypothesis. There is no significant correlation.")
```

As seen on the heatmap before, the **Pearson Correlation Coefficient** of 0.9852 indicates that *there is a strong positive correlation* between the two disorders.
The **p-value** is very small. So we can *reject the null hypothesis* and confirm that there is a relation between prevalances Bipolar disorder and Eating disorders.
